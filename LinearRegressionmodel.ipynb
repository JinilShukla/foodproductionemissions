LINEAR REGRESSION MODEL BY TRAIN-TEST SPLIT APPROACH


```python
#Building a Simple Linear Regression model with Python
#We are going to predict the total emissions of a food by just knowing the emissions occured in the farm
```

```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
```

```python
#Loading the dataset
data=pd.read_csv('Food_Production.csv')
data
```

```python
#scatter plot relation between the two quantites which we are going to compare
data.plot(kind='scatter',x='Farm',y='Total_emissions')
plt.show()
```

```python
#change to dataframe variables
X=pd.DataFrame(data['Farm'])
Y=pd.DataFrame(data['Total_emissions'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=14)
```

```python
#Training the model
```

```python
len(X_train)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.linear_model import LinearRegression
clf=LinearRegression()
```

```python
clf.fit(X_train,Y_train)
```

```python
# Visualising the Training set results
plt.scatter(X_train, Y_train, color = 'red')
plt.plot(X_train, clf.predict(X_train), color = 'blue')
plt.xlabel('Farm Emissions')
plt.ylabel('Total_emissions')
```

```python
# Predicting the Test set results
y_pred = clf.predict(X_test)
```

```python
clf.predict(X_test)
```

```python
# Visualising the Test set results
plt.scatter(X_test, Y_test, color = 'red')
plt.plot(X_train, clf.predict(X_train), color = 'blue')
plt.xlabel('Farm Emissions')
plt.ylabel('Total_emissions')
plt.show()
```

```python
print("R^2: {}".format(clf.score(X_test, Y_test)))
rmse = np.sqrt(mean_squared_error(Y_test, y_pred))
print("Root Mean Squared Error: {}".format(rmse))
```

```python
#hence you can see the regression line spotted
#The R2 score of the model came out to be 0.887331738618805 and the RMSE is 2.4958248137724026 which indicates that the model is highly accurate
```
